### 8)

```python
#Se obtiene la latitud y Longitud de Madrid
Lat_Madrid = ds[(ds['Provincia']=='Madrid') & (ds['Municipio']=='Madrid')]['Y']
Lon_Madrid = ds[(ds['Provincia']=='Madrid') & (ds['Municipio']=='Madrid')]['X']
```

```python
Municipios = ds[ds['Municipio']!='Madrid']

lista = []
latitudes = []
longitudes = []


for i in range(len(Municipios)):
    lista.append(Municipios.iloc[i]['Municipio'])
    latitudes.append(Municipios.iloc[i]['Y'])
    longitudes.append(Municipios.iloc[i]['X'])

data = {'Y' : latitudes, 'X' : longitudes}
df1 = pd.DataFrame(data, index=lista)

Dist1 = lambda x: (x-Lon_Madrid)**2
Dist2 = lambda y: (y-Lat_Madrid)**2

df1['CuadradosX']  = df1['X'].apply(Dist1)
df1['CuadradosY']  = df1['Y'].apply(Dist2)
df1['Distancia'] = np.sqrt(df1['CuadradosX'] + df1['CuadradosY'])*100

print(f'Municipio mas cercano de Madrid: ', df1.sort_values(by='Distancia').head(1).index.values[0])
print(f'Municipio mas lejano de Madrid: ', df1.sort_values(by='Distancia').tail(1).index.values[0])

```
